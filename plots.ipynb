# THIES Data Visualization
A simple group of plots of THIES Data from EPII. Dates between May (2024-05-28) and September (2024-09-30).

```python
from thiesBP import THIESData, THIESDayData
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

folder = './BINFILES/ARCH_AV1'
data = THIESData('av', folder)
data.load_df()
data.complete_empty_dates()
```

```python
def plot_parameter(df: pd.DataFrame, y: str):
    df[y] = pd.to_numeric(df[y], errors='coerce')
    ig, ax = plt.subplots(figsize=(30, 4))
    sns.pointplot(x="Date", y=y, ax=ax, data=df, marker='o', hue=None)
    plt.xlabel('Date')
    plt.ylabel(y)
    plt.title(f'{y} over Time')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Access parameters measured
data.parameters
```

```python
for p in data.parameters:
    plot_parameter(data.dataDF, p)
```
